```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\joy\Desktop\Sea Level\data\csiro_alt_gmsl_mo_2015_csv_edited.csv")
df['id'] = df.index
df.dropna(subset=["GMSL"], inplace=True)
print(df)
```

```python
x = np.array(df.index)
y = np.array(df.GMSL)
```

```python
plt.bar(x, y, color='b')
plt.plot(x, y, color='b')
plt.xlabel("1993-January to 2014-December")
plt.ylabel("Sea Level in mm")
plt.title("Sea Level Rise")
plt.grid(False)
plt.show()
```
